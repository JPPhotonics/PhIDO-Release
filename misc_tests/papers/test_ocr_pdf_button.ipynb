```python
import pytesseract
from IPython.display import display
from PIL import Image, ImageDraw, ImageOps

# Load the image
image = Image.open("Screenshot1.png")
image = ImageOps.grayscale(image)
image = ImageOps.invert(image)
# image = ImageEnhance.Color(image).enhance(0)
# image = ImageOps.invert(image)

# image = image.convert('L').point(lambda x: 0 if x < 128 else 255, '1')


# Use pytesseract to extract data with bounding boxes
data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

# Specify the word you are looking for
word_to_find = "PDF"  # Replace with your word

# Initialize a list to store bounding box details
bounding_boxes = []

# Iterate over all detected words
for i in range(len(data["text"])):
    if word_to_find.lower() in data["text"][i].lower():
        # Get the bounding box coordinates
        x = data["left"][i]
        y = data["top"][i]
        width = data["width"][i]
        height = data["height"][i]
        bounding_boxes.append((x, y, width, height))

# Draw the bounding boxes on the image
draw = ImageDraw.Draw(image)
for bbox in bounding_boxes:
    print("=======>", x, y)
    x, y, width, height = bbox
    draw.rectangle([x, y, x + width, y + height], outline="red", width=2)

# Display the image inline in the Jupyter Notebook
display(image)

# Optionally, save the image with bounding box (optional)
# image.save('output_image_with_bounding_box.png')
```

```python
import easyocr
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw

# Step 1: Initialize EasyOCR Reader
reader = easyocr.Reader(["en"])  # Initialize with English language support

# Step 2: Load the image
image_path = "screenshot---.png"  # Replace with your image path
image = Image.open(image_path)

# Step 3: Perform OCR
results = reader.readtext(image_path)

# Step 4: Specify the word to find
word_to_find = "DOWNLOAD PAPER"  # Replace with the word you're searching for

# Step 5: Search for the word in the OCR results and get its bounding box
bounding_boxes = []
for bbox, text, _prob in results:
    if word_to_find.lower() in text.lower():  # Case-insensitive search
        bounding_boxes.append(bbox)

# Step 6: Draw bounding boxes on the image
draw = ImageDraw.Draw(image)
for bbox in bounding_boxes:
    print("======>", bbox)
    # Convert bbox to a flat list of coordinates (flatten the list)
    flattened_bbox = [tuple(coord) for coord in bbox]
    draw.polygon(flattened_bbox, outline="cyan", width=2)

# Step 7: Display the image with bounding boxes inline in Jupyter Notebook
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis("off")  # Hide axes
plt.show()
```
